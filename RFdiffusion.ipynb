```python
# @title 1. Initialize Foundry & Create JSON Input
from google.colab import drive
import json
import os
drive.mount('/content/drive')
os.environ["FOUNDRY_CHECKPOINT_DIRS"] = "/content/drive/MyDrive/RFD3_Checkpoints"
!pip install "rc-foundry[rfd3]"
!foundry install rfd3
# Create a sample JSON configuration for an active site
# This tells RFD3 to keep specific residues from a PDB (the active site)
# and build a new scaffold around them.
design_config = {
    "inference": {
        "num_designs": 1,
        "output_prefix": "/content/active_site_design",
        "input_pdb": "/content/target.pdb" # You must upload this file!
    },
    "contigmap": {
        "contigs": ["A10-10/A42-42/A80-80/50-100"], # Example: scaffold around 3 specific residues
        "length": "100-150"
    }
}
with open('/content/design_params.json', 'w') as f:
    json.dump(design_config, f, indent=4)
print("Setup complete and design_params.json created.")
```

```python
# @title 2. Run Design from JSON
# Make sure you have uploaded 'target.pdb' to the /content/ folder in Colab first!
!python -m rfd3.inference \
    inputs=/content/design_params.json \
    skip_existing=False
```
